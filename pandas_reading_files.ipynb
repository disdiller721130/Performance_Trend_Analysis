```python
# Dependencies
import pandas as pd
import glob
```

```python
def file_concat(path):
    all_files = glob.glob(path + "/*.csv")
    
    file_list = []
    for filename in all_files:
        overall_df = pd.read_csv(filename, index_col = None, header = 0)
        file_list.append(overall_df)
    
    final_raw_df = pd.concat(file_list, axis = 0, ignore_index = True)
    
    return final_raw_df
```

```python
# Store filepath in a variable
file_path = r"C:\Users\eugene.chu1\Desktop\Data Munging\Raw\OV10650"
combine_data_df = file_concat(file_path)
combine_data_df = combine_data_df.drop(" ", axis = 1)
```
